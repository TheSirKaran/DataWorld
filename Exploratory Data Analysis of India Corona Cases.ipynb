# Corona Case Analysis of India based on States

```python
import pandas as pd
import numpy as np
df=pd.read_csv("IndiaCorona.csv")
```

```python
total_states=np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df=df.iloc[:-1,1:]
```

```python
df
```

# Active Corona Cases In India

```python
import matplotlib.pyplot as plt
fig= plt.figure(figsize=(23,24))
axes= fig.add_axes([0.1,0.5,0.8,0.8])
x=df["state_name"]
y=df["active"]
axes.barh(x,y,label="Corona Deaths",color="magenta")
plt.xlabel("Active Cases")
plt.ylabel("States")
plt.yticks(total_states,df['state_name'])
plt.title("No. Of active Cases")
plt.legend()
plt.show()
```

## Positive Corona Cases In India

```python
import matplotlib.pyplot as plt
fig= plt.figure(figsize=(23,24))
axes= fig.add_axes([0.1,0.5,0.8,0.8])
x=df["state_name"]
y=df["positive"]
axes.barh(x,y,label="Corona Deaths",color="green")
plt.xlabel("Positive Cases")
plt.ylabel("States")
plt.title("No. Of Positive Cases")
plt.legend()
plt.show()
```

#  Cured Corona Cases In India

```python
import matplotlib.pyplot as plt
fig= plt.figure(figsize=(23,24))
axes= fig.add_axes([0.1,0.5,0.8,0.8])
x=df["state_name"]
y=df["cured"]
axes.barh(x,y,label="Corona Deaths",color="blue")
plt.xlabel("Cured Cases")
plt.ylabel("States")
plt.title("No. Of Cured Cases")
plt.legend()
plt.show()
```

# Stacked Corona Cases in India

```python
import matplotlib.pyplot as plt
fig= plt.figure(figsize=(23,19))
axes= fig.add_axes([0.1,0.5,0.8,0.8])
x=df["state_name"]
y=df["cured"]
z=df["death"]
p=df["active"]
q=df["positive"]
plt.barh(x,q,label="positive cases",color="green")
plt.barh(x,y,label="Cured",color="red")
plt.barh(x,p,label="active cases",color="magenta")
plt.barh(x,z,label="Corona Deaths",color="blue")
plt.xlabel("Corona Issues")
plt.ylabel("States")
plt.title("No. Of Death Cases")
plt.legend()
plt.show()
```
